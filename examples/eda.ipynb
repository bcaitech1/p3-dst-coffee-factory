```python
#### LOAD DATA ####

with open('../input/data/train_dataset/train_dials.json', 'r') as f:
    train_dials = json.load(f)
    
with open('../input/data/train_dataset/slot_meta.json', 'r') as f:
    slot_meta = json.load(f)
    
with open('../input/data/train_dataset/ontology.json', 'r') as f:
    ontology = json.load(f)
```

```python
#### DATAFRAME ###

# 1. id
# 2. conversation texts
# 3. domain
# 4. state

idx_lst = [] # 1
dia_lst = [] # 2
dom_lst = [] # 3
sta_lst = [] # 4

for idx in range(len(train_dials)):
    data = train_dials[idx] # current data
    # 1
    idx_lst.append(data['dialogue_idx'])
    # 2
    dialogue = ' '.join([s['text'] for s in data['dialogue']])
    dia_lst.append(dialogue)
    # 3 
    dom_lst.append(data['domains'])
    # 4
    for d in data['dialogue']:
        try:
            state = d['state']
        except:
            continue
    sta_lst.append(state)

data = {'id': idx_lst, 'conversation_texts': dia_lst, 'domain': dom_lst, 'state': sta_lst}

df = pd.DataFrame(data)

del idx_lst
del dia_lst
del dom_lst
del sta_lst
del data

display(df.head())
```

```python
#### DOMAINS ####

domains = []
for lst in df['domain'].to_list():
    domains.extend(lst)
    
dom_val_cnt = Counter(domains)

print(dom_val_cnt)

plt.hist(domains)
plt.show()
```

```python
#### SLOTS ####
# ex) {관광: ['종류', '종류', '지역', ...], ...}

def slots(x):
    slots_lst = [_.split('-') for _ in x['state']]
    slots_dct = dict()
    for lst in slots_lst:
        if lst[0] not in slots_dct:
            slots_dct[lst[0]] = [lst[1]]
        else:
            slots_dct[lst[0]].append(lst[1])
    return slots_dct

slots_processed = df.apply(lambda x: slots(x), axis=1)

slots_merged = dict()

for row in slots_processed:
    for key, val in row.items():
        if key not in slots_merged:
            slots_merged[key] = [_ for _ in val]
        else:
            slots_merged[key].extend(val)
            
for key in dom_val_cnt.keys():
    print(key)
    print(Counter(slots_merged[key]))
    print()
    plt.xticks(rotation=90)
    plt.hist(slots_merged[key])
    plt.show()
```

```python
#### N-GRAMS PER DOMAIN ####

def ngrams_dcts(dials, N):
    gram_dct = {k: [] for k in dom_val_cnt.keys()} # Save nGram data per domain
    # iterates data
    for idx in range(len(dials)):
        prev_state = []
        for s in dials[idx]['dialogue']:
            if s['role'] != 'user': # only user queries
                continue
            cur_state = s['state']
            added_state = list(set(cur_state) - set(prev_state))
            if len(added_state) != 0:
                domains = list(set([_.split('-')[0] for _ in added_state]))
                for cur_dom in domains:
                    words = s['text'].split(' ')
                    if len(words) < N:
                        continue
                    for idx_ in range(len(words)):
                        if idx_ == len(words) - N + 1:
                            break
                        next_gram = tuple([_ for _ in words[idx_: idx_ + N]])
                        gram_dct[cur_dom].append(next_gram)
            prev_state = cur_state
    return gram_dct
```

```python
#### BIGRAM ####
#### QUERIES ONLY #####

print('==============BIGRAM===============', '\n')
bigram_dct = ngrams_dcts(train_dials, 2)
TOP_N = 10
for dom in dom_val_cnt.keys():
    print(f'============ {dom} TOP{TOP_N} ============')
    pprint(Counter(bigram_dct[dom]).most_common(TOP_N))
    print()
```

```python
#### TRIGRAM ####
#### QUERIES ONLY #####

print('==============TRIGRAM==============', '\n')
trigram_dct = ngrams_dcts(train_dials, 3)
TOP_N = 10
for dom in dom_val_cnt.keys():
    print(f'============ {dom} TOP{TOP_N} ============')
    pprint(Counter(trigram_dct[dom]).most_common(TOP_N))
    print()
```
